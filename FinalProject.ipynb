### Introduction

We have three datasets provided by WPRDC: Arrest Rate Dataset, Playground Dataset, and Community Engagement Dataset. We have to analyze all the datasets containing information about the neighborhood, and come up with a metric to determine which neighborhood is the best one. In order for us to determine which neighborhood is the best, we tried to think of necessities in a neighborhood that attracts people. For example, people care about the liveliness of a neighborhood, safety, and a place where their children can play. Based on all of the needs, we would determine which neighborhood is mostly likely to satisfy people’s needs. 

### Metric

We thought that a neighborhood that satisfied people’s needs determined the best neighborhood in the datasets. A place with safety, playgrounds for kids, and how engaged the community would help us understand which is the best neighborhood. We focused on three qualities. Safety was one of them; we measured safety by finding out which neighborhood had the least amount of arrests. The one with the least amount meant the neighborhood is the safest. Playgrounds were another quality; we measured the number of playgrounds by determining the neighborhood with the most amount of playgrounds. The more playgrounds, the more kids can play and engage with the community. Lastly, community engagement was another we measured; we determined community engagement by checking the attendance of each neighborhood. The neighborhood with a higher attendance determined how well-engaging the community is. Overall, the metric we decided to use in order to measure "bestness" is Community Engagement Score. The metric reflects how safe, lively, and involved a neighborhood is based on the submetrics from the following three datasets.


### Description of Datasets

- Arrest Rate Dataset: This dataset shows information about arrests that were made in each neighborhood. 
- Playgrounds Dataset: This dataset portrays the number of playgrounds in each neighborhood.
- Community Engagement Dataset: This dataset displays the attendance in each community.
